# Q1: What is Dimensionality Reduction? Why is it important in machine learning?

"""
Dimensionality Reduction is the process of reducing the number of input features (variables)
while preserving the essential information or patterns in the data.

Machine learning models often struggle with high-dimensional data due to:
- Increased computational cost
- Risk of overfitting
- Difficulty in visualization

Hence, reducing the number of features helps in:
1. **Improving Model Performance** â€“ Fewer irrelevant features mean better accuracy and generalization.
2. **Reducing Overfitting** â€“ Less complexity reduces noise in training data.
3. **Enhancing Visualization** â€“ Makes it possible to plot data in 2D/3D.
4. **Faster Computation** â€“ Models train and predict faster with fewer dimensions.

ðŸ“˜ Example:
Suppose you have a dataset with 100 features describing customer behavior.
Only 10 of them actually influence customer churn.
Dimensionality reduction helps you keep those 10 most informative features.

# Q2: Name and briefly describe three common dimensionality reduction techniques.

"""
1. **Principal Component Analysis (PCA)**:
   - Linear method that transforms data into new axes (principal components) capturing maximum variance.
   - Reduces correlated features.
   - Example: Converting 10 correlated financial ratios into 2 uncorrelated components.

2. **Linear Discriminant Analysis (LDA)**:
   - Supervised method that maximizes class separation.
   - Often used in classification tasks like face recognition.
   - Example: Reducing 50 facial features to 2 dimensions that separate male vs female faces.

3. **t-Distributed Stochastic Neighbor Embedding (t-SNE)**:
   - Non-linear technique ideal for visualizing high-dimensional data in 2D or 3D.
   - Preserves local structure of data.
   - Example: Visualizing 784-dimensional MNIST digit images in 2D.
"""

ðŸ“˜ Example:
Grouping customers by spending patterns or credit card usage.

```python
# Q8: Write Python code to perform K-means clustering on a sample dataset.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Generate synthetic data
X, y = make_blobs(n_samples=200, centers=3, random_state=42, cluster_std=1.5)

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Plot clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=40)
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
# Q9: What is inheritance in OOP? Provide a simple example in Python.

"""
Inheritance allows one class (child) to inherit properties and methods from another (parent),
reducing redundancy and improving reusability.

Types of Inheritance:
1. Single â€“ One parent, one child
2. Multiple â€“ Multiple parents
3. Multilevel â€“ Chain of inheritance

ðŸ“˜ Example:
"""

class Animal:
    def __init__(self, name):
        self.name = name

    def sound(self):
        return "Some generic sound"

class Dog(Animal):
    def sound(self):
        return f"{self.name} says: Woof!"

dog1 = Dog("Tommy")
print(dog1.sound())

"""
Here, `Dog` inherits from `Animal` and overrides the `sound()` method.
This demonstrates **method overriding** and reusability.
"""
```

```python
# Q10: How can time series analysis be used for anomaly detection?

"""
In time series, anomalies are data points that deviate significantly from
expected temporal behavior.

Methods:
1. **Forecasting Models** (ARIMA, Prophet):
   - Predict expected future values. If actuals deviate beyond threshold, mark as anomaly.

2. **Rolling Statistics**:
   - Compare moving averages and standard deviations to detect sudden spikes.

3. **Residual Analysis**:
   - Residual = Actual â€“ Predicted; large residuals indicate anomalies.

ðŸ“˜ Example:
If a store normally sells 100â€“120 units per day, and one day it sells only 10,
time series anomaly detection flags it as unusual.
"""

# Simple Example Code:
import pandas as pd
import numpy as np

# Create synthetic time series
np.random.seed(42)
time = pd.date_range(start='2024-01-01', periods=50)
values = np.random.normal(100, 5, size=50)
values[30] = 160  # Anomaly

series = pd.DataFrame({'Date': time, 'Value': values})

# Detect anomaly using z-score
series['Zscore'] = (series['Value'] - series['Value'].mean()) / series['Value'].std()
series['Anomaly'] = np.where(abs(series['Zscore']) > 2.5, 'Yes', 'No')

print("Detected anomalies:")
print(series[series['Anomaly'] == 'Yes'])

plt.plot(series['Date'], series['Value'], label='Value')
plt.scatter(series.loc[series['Anomaly'] == 'Yes', 'Date'],
            series.loc[series['Anomaly'] == 'Yes', 'Value'],
            color='red', label='Anomaly')
plt.legend()
plt.title("Time Series Anomaly Detection Example")
plt.show()
```
